```python
from pathlib import Path

project_dir = Path(r"C:\ManzaRotation")
raw_dir = project_dir / "Raw"
derivatives_dir = project_dir / "Derivatives"

subject = "sub-100"
subject_deriv_dir = derivatives_dir / subject
subject_raw_dir = raw_dir / subject

tasks = ["modulate1", "modulate2"]

for task in tasks:
    print("=" * 80)
    print(f"TASK: {task}")
    print("=" * 80)

    bold_files = sorted(
        subject_deriv_dir.rglob(
            f"{subject}_task-{task}_space-MNI152NLin2009cAsym_res-2_desc-preproc_bold.nii.gz"
        )
    )

    mask_files = sorted(
        subject_deriv_dir.rglob(
            f"{subject}_task-{task}_space-MNI152NLin2009cAsym_res-2_desc-brain_mask.nii.gz"
        )
    )

    confounds_files = sorted(
        subject_deriv_dir.rglob(
            f"{subject}_task-{task}_desc-confounds_timeseries.tsv"
        )
    )

    events_files = sorted(
        subject_raw_dir.rglob(
            f"{subject}_task-{task}_events.tsv"
        )
    )

    print("\nBOLD:")
    for f in bold_files:
        print(f"  {f}")

    print("\nMASK:")
    for f in mask_files:
        print(f"  {f}")

    print("\nCONFOUNDS:")
    for f in confounds_files:
        print(f"  {f}")

    print("\nEVENTS:")
    for f in events_files:
        print(f"  {f}")

    print()
```

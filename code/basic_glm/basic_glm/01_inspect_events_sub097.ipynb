```python
from pathlib import Path
import pandas as pd
```

```python
analysis_root = Path(r"C:\ManzaRotation\Analysis")
raw_func = Path(r"C:\ManzaRotation\Raw") / subject / "func"
out_dir = analysis_root / "outputs" / subject / "basic_glm"

out_dir.mkdir(parents=True, exist_ok=True)
```

```python
fb_to_feel = {
    "fb_v_pos": "positive_valence",
    "fb_v_neg": "negative_valence",
    "fb_a_pos": "positive_arousal",
    "fb_a_neg": "negative_arousal",
}
```

```python
def make_feel_only_events(events):
    events_labeled = events.copy()
    events_labeled["trial_type_labeled"] = events_labeled["trial_type"]

    for idx in events_labeled.index:
        trial_type = events_labeled.loc[idx, "trial_type"]

        if trial_type == "feel":
            prev_type = events_labeled.loc[idx - 1, "trial_type"]
            events_labeled.loc[idx, "trial_type_labeled"] = fb_to_feel[prev_type]

    glm_events = events_labeled[events_labeled["trial_type"] == "feel"].copy()
    glm_events = glm_events[["onset", "duration", "trial_type_labeled"]].copy()
    glm_events = glm_events.rename(columns={"trial_type_labeled": "trial_type"})

    return events_labeled, glm_events
```

```python
# Modulate 1
events_path1 = raw_func / f"{subject}_task-modulate1_events.tsv"
events1 = pd.read_csv(events_path1, sep="\t")

events1_labeled, glm_events1 = make_feel_only_events(events1)

out_path1 = out_dir / f"{subject}_task-modulate1_events-feelOnly.tsv"
glm_events1.to_csv(out_path1, sep="\t", index=False)

out_path1.exists()
```

```python
# Modulate 2
events_path2 = raw_func / f"{subject}_task-modulate2_events.tsv"
events2 = pd.read_csv(events_path2, sep="\t")

events2_labeled, glm_events2 = make_feel_only_events(events2)

out_path2 = out_dir / f"{subject}_task-modulate2_events-feelOnly.tsv"
glm_events2.to_csv(out_path2, sep="\t", index=False)

out_path2.exists()
```
